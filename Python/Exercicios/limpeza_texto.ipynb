```python
palavras = [p.lower().strip(string.punctuation) for p in texto.split()]
palavras = [p.replace('‘','') for p in palavras]
palavras = [p.replace('“','') for p in palavras]
palavras = [p.replace('”','') for p in palavras]
```

```python
freqdist = {}
for palavra in palavras:
    if palavra in freqdist:
        freqdist[palavra] += 1
    else:
        freqdist[palavra] = 1
freqdist
```
