```python
#seu código aqui
#a)
def freqdict(texto):
    freqdict = {}
    palavras = texto.split()
    palavras = [word.replace('\ufeff','') for word in palavras] #só para ajeitar o começo do texto
    palavras = [word.replace('«','') for word in palavras]
    palavras = [word.replace('»','') for word in palavras]
    palavras = [word.strip(string.punctuation) for word in palavras]
    palavras = [word.lower() for word in palavras]
    palavras = [word for word in palavras if word != '']
    for word in palavras:
        if word in freqdict.keys():
            freqdict[word] += 1
        else:
            freqdict.update({word:1})
    return freqdict
letra_a = freqdict(texto)
letra_a
```

Questão 2 (2.0 pontos)  

Seja d(n) definido como a suma dos divisores inteiros de n. Se d(a) = b e d(b) = a, sendo a ≠ b, então a e b são um "par amigável" de números.  

Por exemplo, os divisores inteiros de 220 são 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 e 110; então d(220) = 284.  
Os divisores inteiros de 284 são 1, 2, 4, 71 e 142; então d(284) = 220.  

Encontre todos os "pares amigáveis" de números menores que 20000.  Crie uma lista de tuplas com esses pares.  

```python
#seu código aqui
def somad(n):
    divisores = [1]
    for tent in range(2,int(n/2)+1):
        if n%tent == 0:
            divisores.append(tent)
    somad = 0
    for div in divisores:
        somad += div
    return somad #essa funcao ta lerda
```

```python
amigos = []
for i in range(10,20000):#sabendo que antes de 10 não tem
    j = somad(i)
    lista_jacont = []
    if (somad(j) == i) and (i != j) and (i not in lista_jacont):
        amigos.append((i,j))
        lista_jacont.append(j)
amigos
```

Usando funções geradoras, crie um objeto que emule um dado de oito faces (1-8) que seja viciado, de tal forma que a probabilidade desta função geradora retornar certo valor seja proporcional ao próprio valor (por exemplo, a face 6 é 3 vezes mais provável de sair do que a face 2).  

(a) Usando os comandos plt.plot ou plt.hist, mostre graficamente o resultado de 10000 lançamentos;  
(b) Modifique esta função geradora de forma que ela termine automaticamente quando todos os valores possíveis (1,2,3,4,5,6,7,8) tiverem sido retornados pelo menos uma vez. Neste caso, ela retornará o tempo absoluto total que ocorreu desde a primeira iteração.  

```python
#seu código aqui
def dado_vic():
    dado = ''
    for i in range(1,9):
        dado += i*str(i)
    lista_dado = [int(num) for num in dado]
    while True:
        indice = random.randint(0,len(lista_dado)-1)
        yield lista_dado[indice]
```

```python
gen_results = dado_vic()
```

```python
next(gen_results)
```

```python
gen_results = dado_vic()
results = []
for i in range(10000):
    gen_results = dado_vic()
    r_dado = next(gen_results)
    results.append(r_dado)
freq = np.zeros(8)
for dd in results:
    freq[dd-1] += 1
x = np.arange(1,9)
plt.bar(x,freq)
plt.xlabel('nº do dado')
plt.ylabel('frequência')
plt.show() #cresc. aprox. linear com o cresc do número do dado
```
